# Bar Charts

### In this notebook we learn and practice plots used in statistics step by step.

Bar charts are used for the graphical represantation for values of quantitative / categorical data.
<br>
In a bar chart there are rectangular columns, that usually stand for the frequency (times of occurrence) or relative frequency (times of occurrence divided by sample size).
<br>
A bar chart can be vertical or horizontal.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# blood group frequency bar chart (sample size = 40)

categories = ['O', 'A', 'B', 'AB']
values = [16, 12, 8, 4]

# Create a simple bar chart
plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.xlabel('Blood Groups')
plt.ylabel('Frequency v')
plt.title('Frequency Distribution of Blood Groups')
plt.show()




```

```python
# blood group relative frequency bar chart (sample size = 40)

# now we customize the bar chart a little
# we change the width of columns and rotate the name of each group 

values = [.4, .3, .2, .1]

plt.bar(categories, values, color='skyblue', width=0.6, edgecolor='black')
plt.xlabel('Blood Groups')
plt.ylabel('Rerative Frequency f')
plt.title('Relative Frequency Distribution of Blood Groups')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()


```

```python
# now let's plot both of the charts simultaneously and horizontally

# for the first bar chart
plt.subplot(2, 1, 1)
plt.barh(["O", "A", "B", "AB"], [16, 12, 8, 4], color='purple', edgecolor='black')
plt.xlabel('Frequency v')
plt.ylabel('Blood Groups')
plt.title('Frequency Distribution of Blood Groups')

# for the second bar chart
plt.subplot(2, 1, 2)
plt.barh(["O", "A", "B", "AB"], [.4, .3, .2, .1], color='darkcyan', edgecolor='black')
plt.xlabel('Rerative Frequency f')
plt.ylabel('Blood Groups')
plt.title('Relative Frequency Distribution of Blood Groups')

plt.tight_layout()
plt.show()
```

```python
# blood group and Rhesus relative frequency 

# Data for the groups
Groups = ['O', 'A', 'B', 'AB']
RhPos = [14, 10, 7, 3]
RhNeg = [2, 2, 1, 1]

# range of values for the X-axis
X_axis = np.arange(len(Groups))


plt.bar(X_axis - 0.2, RhPos, 0.4, label='Rh (+)', color="#156082")
plt.bar(X_axis + 0.2, RhNeg, 0.4, label='Rh (-)', color="#A02B93")


plt.xticks(X_axis, Groups)
plt.xlabel("Blood Groups")
plt.ylabel("Frequency v")
plt.title("Rhesus Blood Group Frequency")
plt.legend()
plt.show()
```

```python
# rhesus relative frequncy compare

Groups = ['']
Pos = [0.85]
Neg = [0.15]

# range of values for the X-axis
X_axis = np.arange(len(Groups))


plt.bar(X_axis - 0.1, Pos, 0.2, label='Rh (+)', color="#44B3E1")
plt.bar(X_axis + 0.1, Neg, 0.2, label='Rh (-)', color="#D86DCD")


plt.xticks(X_axis, Groups)
plt.xlabel("Blood Groups")
plt.ylabel("Relative Frequency f")
plt.title("Rhesus Relative Frequency")
plt.legend()
plt.show()
```
